```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df.sample(5)

```

```python
import matplotlib.pyplot as plt
```

```python
df['Survived'].value_counts().plot(kind='bar')
```

```python
plt.bar(df['Sex'],df['Fare'],color='green',alpha=0.5)
```

```python
fig=plt.figure(figsize=(4,5))
plt.bar(df['Sex'],df['Fare'],color='green',alpha=0.5)
```

```python
plt.scatter(df['Age'],df['Fare'],color='red',alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Fare')
```

```python
plt.stem(df['Age'],df['Fare'],linefmt='grey',markerfmt='D',basefmt='black')
plt.xlabel('Age')
plt.ylabel('Fare')
```

```python
plt.stackplot(df['Age'],df['Fare'],colors=['red','green','blue'])
plt.xlabel('Age')   
plt.ylabel('Fare')
```

```python
plt.stairs(df['Fare'],df['Age'],orientation='vertical')
plt.xlabel('Age')
plt.ylabel('Fare')              
plt.show()
```
